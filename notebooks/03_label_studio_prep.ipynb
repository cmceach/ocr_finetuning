# Prepare Data for Label Studio

This notebook prepares documents for annotation in Label Studio, optionally using Azure DI for pre-annotation.


```python
from src.preprocessing.azure_di_preprocessor import AzureDIPreprocessor
import json
```

```python
# Initialize preprocessor
preprocessor = AzureDIPreprocessor()

# Pre-annotate documents
document_paths = ["../data/raw_documents/doc1.pdf", "../data/raw_documents/doc2.pdf"]
image_urls = ["http://localhost:8080/data/upload/doc1.pdf", "http://localhost:8080/data/upload/doc2.pdf"]

preannotations = preprocessor.preannotate_batch(document_paths, image_urls)

# Save for Label Studio import
with open("../data/labeled_data/preannotations.json", "w") as f:
    json.dump(preannotations, f, indent=2)

print(f"Pre-annotated {len(preannotations)} documents")
```
