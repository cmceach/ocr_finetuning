# Data Analysis and Quality Checks

This notebook performs exploratory data analysis on the OCR dataset.


```python
import json
from pathlib import Path
import matplotlib.pyplot as plt
from src.data.label_studio_loader import LabelStudioLoader
```

```python
# Load Label Studio export
json_path = "../data/labeled_data/export.json"
loader = LabelStudioLoader(json_path)
tasks = loader.get_tasks()

print(f"Total tasks: {len(tasks)}")
```

```python
# Analyze text region statistics
region_counts = []
text_lengths = []

for task in tasks:
    regions = loader.extract_text_regions(task)
    region_counts.append(len(regions))
    for region in regions:
        text_lengths.append(len(region.get("text", "")))

print(f"Average regions per image: {sum(region_counts) / len(region_counts):.2f}")
print(f"Average text length: {sum(text_lengths) / len(text_lengths):.2f}")
```
